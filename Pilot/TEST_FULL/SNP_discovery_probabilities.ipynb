# Predicting SNP discovery in Alaskan source populations

***(When reading this notebook, you can focus on the text and figures, no need to understand the code)*** 

The goal of this sequencing is to identify as many loci as possible in the samples of source populations that were introduced to recipient lakes. 

The problem is, we live in a finite world, and sequencing is not unlimited. Thus, while the per-sample coverage required to identify **every single allele** will be around 50x, the sequencing that we are able to do will be less than that. This means that we will inevitably miss alleles. 

Thus the important question is, how many alleles will we miss for a given coverage? 

To answer this question, we need to know several things:

    1) What is the SNP discovery rate for different coverages?  
    2) What is the alelle frequency spectrum for a given population?


### 1) SNP discovery rate

The probability of finding a SNP at least once in a sample of a population can be estimated by calculating the probability using a sample without replacement strategy, based on the frequency and size of the sample. Here we essentially calculate the chance of finding the only the major (highest frequency) allele for a polymorphic locus in the sample and then subtract this from 1 to give the probability of finding the minor allele "at least once".

***Eq. 1)***  
$P$(SNP) $= 1 - \prod_{i = 0}^{N-1} \frac{(REF COUNT-i)}{(REF COUNT -i) + (ALT\ COUNT)}$

***where***
$N$ is the sample size and $q$ is the frequency of the minor SNP allele in the whole population. 
$REFCOUNT$ is the number of the major allele in the population
$ALTCOUNT$ is the number of the minor allele

***However** unfortunately we can't know the allele counts in the population, only in our sample. 

For that reason we have to assume an infinite population size, and thus not worry about replacement. Hopefully <i> G. aculeatus </i> are large enough that this isn't a problem.

We can therefore use this much simpler equation (**Note it also assumes that things are at hardy weinberg**). 

***Eq. 2)***  
$P_{SNP}$ = $1 - (1 - q)^N$

where $N$ is the size of the sample and $p$ is the frequency of the minor (rare) allele at a locus

Using this equation, we can estimate some probabilities for SNP discovery at different allele frequencies.



### What is the probability of finding a SNP at a given allele frequency and sample size?


```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def find_Psnp(N, q):  ## this function corresponds to the Eq. 2) above. 
   
    P = 1-pow((1 - q),N)
    return P
```

```python
qs = [i/10000 for i in range(0, 220,20)][1:] ## testing for a range of minor allele frequencies between 0.002 - 0.02
Ns = range(100, 1000, 100)  ## and for a sample size of between 100 - 1000

results = {}

for q in qs:
    results[q] = []
    for N in Ns:
        Psnp = find_Psnp(N, q)
        results[q].append(Psnp)
        
plt.figure(figsize = (10,7))

for q in results:
    plt.plot(Ns, results[q], label=q)
plt.legend(title = "q")
plt.title("Figure 1. Probability of finding SNP")
plt.xlabel("Sample size (N)")
plt.ylabel("P(SNP)")
plt.show()
```

### Accounting for imperfect allele sampling in individuals

***Eq. 2) assumes that, if we sample an individual we have a probability = 1 of finding all alleles in that sample. In practice this is not true. ***

So how likely are we to find an allele in a given sample? 

We can modify the above equation to include this. 

***Eq. 3)*** $P_{SNP}$ = $1 - (1 - (q*P_{IND}))^N$

Here I multiply the probability of sampling the minor allele in each individual by the probability of finding the allele if it is present. 

Lets see how this would (theoretically) change the above plot. To do that we focus on just one allele frequency.  


```python
def find_Psnp_Pind(N, q, Pind):
    
    P = 1 - pow((1 - (q*Pind)),N)
    return P
```

```python
qs = [i/10000 for i in range(0, 220,20)][1:]
q = 0.01 ## using a frequency of 0.01
Ns = range(100, 1000, 100)
Pinds = [i/10 for i in range(1,11,1)]

Pind_results = {}

for Pind in Pinds:
    Pind_results[Pind] = []
    for N in Ns:
        Psnp = find_Psnp_Pind(N, q, Pind)
        Pind_results[Pind].append(Psnp)
```

```python
plt.figure(figsize = (10,7))

for Pind in Pind_results:
    plt.plot(Ns, Pind_results[Pind], label=Pind)
plt.legend(title = "Pind")
plt.title("Figure 2. Probability of finding SNP (q = %s)" % q)
plt.xlabel("Sample size (N)")
plt.ylabel("P(SNP)")
plt.show()
```

So, with low chance of finding an allele in a given sample $P_{IND}$, we have quite low chance of finding an allele until sample sizes get very high

### What is $P_{IND}$ in our dataset?

We need to know what $P_{IND}$ is for our various coverages and sample sizes before we can decide what value is acceptable? 

We can calculate this using our pilot data. To do this, I will first extract a sets of SNPs of different allele frequencies from the full pilot dataset (38x). I will then check to see if these SNPs are present in the dataset for the coverage downsamples. The proportion of them that are present will be our proxy for the proability of detecting an allele at that depth. 

**Note, we can do this because we know, from the coverage downsampling analyses, that our SNP discovery was plateauing at high coverages (38x). Thus we can assume that $P_{IND}$ for our pilot data is approaching 1, and at 38x, our allele sampling should be similar to the first figure, in which we have good power to detect alleles with frequencies as low as 0.002 in the population (as N was almost 200).*** 



```python
## Get SNP IDs from the 38x pilot SNP set.

freq_paths = "/home/djeffrie/Data/Stickleback/FITNESS/MCGILL_Test/1.0/SL_frequencies.frq"

kept_freqs = {}

with open(freq_paths) as frequencies:
    
    for line in frequencies:
        
        if not line.startswith("CHROM"):
            freq = float(line.split()[5].split(":")[1])
        
            if 0.02 <= freq < 0.05:  ## note, there are 19 samples in this population, so the minimum frequency is 0.0263
                ID = "_".join([line.split()[0], line.split()[1]])
                kept_freqs[ID] = freq
                
```

I did the same for several other allele frequency ranges, and below are the values for $P_{IND}$ for all. I also fit a polynomial here so we can extrapolate for intermediate coverages. 

```python
depths = [38, 30, 23, 15]

## 0.0263 < q < 0.05
Pinds_005 = [1,0.96,0.85,0.28]

z_005 = np.polyfit(depths, Pinds_005, 3) # calculate polynomial
f_005 = np.poly1d(z_005)
depths_poly_005 = np.linspace(depths[0], depths[-1], 50) # calculate new x's and y's
Pinds_poly_005 = f_005(depths_poly_005)

## 0.05 < q < 0.1
Pinds_01 = [1,0.96,0.86,0.31]

z_01 = np.polyfit(depths, Pinds_01, 3) # calculate polynomial
f_01 = np.poly1d(z_01)
depths_poly_01 = np.linspace(depths[0], depths[-1], 50) # calculate new x's and y's
Pinds_poly_01 = f_01(depths_poly_01)

## 0.1 < q < 0.2
Pinds_02 = [1,0.97,0.87,0.37]

z_02 = np.polyfit(depths, Pinds_02, 3) # calculate polynomial
f_02 = np.poly1d(z_02)
depths_poly_02 = np.linspace(depths[0], depths[-1], 50) # calculate new x's and y's
Pinds_poly_02 = f_02(depths_poly_02)

# 0.2 < q < 0.3
Pinds_03 = [1,0.97,0.90,0.47]

z_03 = np.polyfit(depths, Pinds_03, 3) # calculate polynomial
f_03 = np.poly1d(z_03)
depths_poly_03 = np.linspace(depths[0], depths[-1], 50) # calculate new x's and y's
Pinds_poly_03 = f_03(depths_poly_03)

# 0.3 < q < 0.4
Pinds_04 = [1,0.98,0.91,0.54]

z_04 = np.polyfit(depths, Pinds_04, 3) # calculate polynomial
f_04 = np.poly1d(z_04)
depths_poly_04 = np.linspace(depths[0], depths[-1], 50) # calculate new x's and y's
Pinds_poly_04 = f_04(depths_poly_04)

plt.figure(figsize = (10,7))

plt.scatter(depths, Pinds_005)
plt.scatter(depths, Pinds_01)
plt.scatter(depths, Pinds_02)
plt.scatter(depths, Pinds_03)
plt.scatter(depths, Pinds_04)

plt.plot(depths_poly_005, Pinds_poly_005, label = "0.02 - 0.05")
plt.plot(depths_poly_01, Pinds_poly_01, label = "0.05 - 0.1")
plt.plot(depths_poly_02, Pinds_poly_02, label = "0.1 - 0.2")
plt.plot(depths_poly_03, Pinds_poly_03, label = "0.2 - 0.3")
plt.plot(depths_poly_04, Pinds_poly_04, label = "0.3 - 0.4")

plt.legend(title = "q")
plt.ylabel("Pind")
plt.xlabel("Depth")
plt.show()
```

So we can see that the probability of discovering a given SNP drops off rapidly below 25x for all allele frequency ranges. However above that, SNP discovery rates are pretty good. We can also see that $P_{IND}$ does not differ that much for different allele frequencies, for example, to achieve a $P_{IND}$ = 0.7, a depth of between ~17-20x is required for all frequency ranges. 

We now have $P_{IND}$, so we can calculate the probability of finding alleles for different frequencies in a sample. 

## Estimating SNP discovery rates in our data

So we know the sample size will be around 1000 for each popualtion. And we now know how coverage affects SNP discovery in individuals. So we can now find the probability of discovering an allele at a given frequency. 

```python
qs = [i/10000 for i in range(0, 110,10)][1:]
depths = range(15,40,1)
N = 1000

## using the polynomials to find Pind
Pind_poly_dict = {}
Pind_poly_dict[(0.00,0.05)] = f_005  ## I'm using this function for the lowest freqs. But note there was no data frequencies 0.023 in the pilot data, so this is extrapolation. The models don't differ that much though, so I think its safe
Pind_poly_dict[(0.05,0.1)] = f_01
Pind_poly_dict[(0.1,0.2)] = f_02
Pind_poly_dict[(0.2,0.3)] = f_03
Pind_poly_dict[(0.3,0.4)] = f_04

results = {}

for q in qs:
    results[q] = []
    for qrange in Pind_poly_dict:
        if qrange[0] < q <= qrange[1]:
            
            for depth in depths:
                Pind = Pind_poly_dict[qrange](depth)
                Psnp = find_Psnp_Pind(N, q, Pind)
                results[q].append(Psnp)
```

```python
plt.figure(figsize = (10,7))

for q in results:
    plt.plot(depths, results[q], label=q)
plt.legend(title = "q")
plt.title("Probability of finding SNP in the population sample (N = %s)" % N)
plt.xlabel("Depth")
plt.ylabel("P(SNP)")
plt.show()
```

## Summary

So, the good news is, if we are to trust the above calculations, our sample sizes should be high enough that they mostly mitigate the loss of power due to coverage. Depth only starts to matter at very low allele frequencies, when alleles are present in only a few samples. 

With $N=1000$, and effective coverage of >15x only alleles with frequencies below $0.005$ become problematic to detect. To put that into context, in a sample of 1000, an allele at a frequency of $q = 0.001$ is present in 2 copies in the sample, an allele with $q = 0.005$ is present in 10, and an allele with $q = 0.01$ is present in 20 individuals. Hopefully missing some such loci will not be too problematic. 

So from this, my feeling is that coverage is not as impactful on SNP discovery as we initially thought. And when it is, it is only for extremely low frequency alleles. 

In light of this, I think the best strategy for sequencing is to prioritise the number of samples, and just push for the highest coverage possible. However, I would caution that below 15x ***effective coverage*** we are getting into dangerous territory again even for alleles of higher frequency, as discovery rates drop off very sharply at the bottom of the depth range (<15x). So I would advise a raw coverage of at least 20 as a minimum.

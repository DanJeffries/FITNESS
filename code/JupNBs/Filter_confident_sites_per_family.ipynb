# Filtering confident sites

As part of the process of creating the training data for the DV training I need to identify loci that are confidently non-variant in each family. These will be used to show the model what non-variant positions look like, so it can better identify false positives. 

Importantly, because the training will be done on the offspring, we do not want to use evidence from the offspring in the training. E.g. if there are a couple of reads that erroneously map in the offspring, but we know, based on confident calls in the parents that this is not a real SNP, then this is very useful training data! Thus, we will only use evidence from the parents on whether a locus in the offspring should be non-variant. 

So . . . the criteria is that loci should be kept if the parents are both homozygous FOR THE SAME ALLELE

```python
import gzip as gz

father_index = 0
mother_index = 1
offspring_index = 2

index = 0
removed = []
kept_sites = []
with gz.open("/Users/dj20y461/Downloads/FG_parents.g.vcf.gz", "rt") as vcf:
    
    for line in vcf:
        
        line = line.strip()

        if not line.startswith('#'):
            
            chrom = line.split()[0]
            pos = line.split()[1]
            GT_fields = line.split()[9:]
                
            site_GTs = []
                
            for sample in GT_fields:
                site_GTs.append(sample.split(":")[0])

            if site_GTs[father_index] == site_GTs[mother_index]:
                        
                kept_sites.append((chrom, pos, site_GTs))
                    
            else:
                removed.append((chrom, pos, site_GTs))
                        
            index += 1
                
            if index % 1000000 == 0:
                print(index, "sites processed")
                
        else:
            pass 
            
            
        
        
```
